```python
import pandas as pd
```

```python
df = pd.read_csv('vivareal_aluguel.csv', dtype=str)
df.shape
```

```python
df.drop_duplicates(keep='first',inplace=True)
```

```python
df.shape
```

```python
df = df.drop(['Unnamed: 0'], axis=1) # comando usado para remover coluna indesejada
```

```python
df['banheiro'] = df['banheiro'].str.extract('(.*) Banheiro')
df.head(3)
```

```python
df['vaga'] = df['vaga'].str.extract('(.*) Vaga')
df.head(3)
```

```python
df['quartos'] = df['quartos'].str.extract('(.*) Quarto')
df['aluguel'] = df['aluguel'].str.extract('R\$ (.*) /Mês')
df.head(3)
```

```python
df['aluguel'] = df['aluguel'].str.replace('\.','',regex=True)
df.head(3)
```
